```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
!sudo apt install tesseract-ocr
!pip install pytesseract
import pytesseract
import shutil
import os
import random
from google.colab import files

#uploaded = files.upload()
#########read the image
img = cv2.imread("/content/sample_data/a01-000u.png")

#########show img
print("AFFICHAGE DE LORDANNACE SOUS IMAGE")
#cv2_imshow(img)
#########croped image

crop_img = img[369:587, 377:2177]
#cv2_imshow(crop_img)
## Binary Thresholding
ret, img_binary = cv2.threshold(crop_img, 127, 255, cv2.THRESH_BINARY)
print("AFFICHAGE DE LORDANNACE SOUS TEXT BINAIRE")
cv2_imshow(img_binary)
print("affiche les dimensions de la matrice:", crop_img.shape)              # affiche les dimensions de la matrice
print("accède à la valeur du premier pixel :", crop_img[0,0])               # accède à la valeur du premier pixel
#convertire image en chaine de caractaire 
img2char = pytesseract.image_to_string(img_binary)
print("CONVERTIRE LORDANNACE SOUS TEXT")
print(img2char)
#convertire image en Beinair 
imgbox = pytesseract.image_to_boxes(img_binary)
print("CONVERTIRE CHAQUE LETTRE AVEC (X, Y, W largeur, H hateur ) ")
print(imgbox)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
